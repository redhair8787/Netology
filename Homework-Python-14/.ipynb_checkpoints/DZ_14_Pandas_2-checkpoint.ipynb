Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
# Загружаем библиотеку
import pandas as pd
```

```python
# Считываем файл
ratings=pd.read_csv('ratings.csv', sep=',')
ratings.head()
```

```python
# Группируем данные по userId
ratings_group = ratings.groupby('userId').count().reset_index()

# Оставляем два столбца 
ratings_group=ratings_group[['userId','rating']]
ratings_group.head()
```

```python
# Оставляем тех, кто выставил более 100 оценок
ratings_group_more_than_100 = ratings_group[ratings_group['rating']>100]

# Переименовываем столбец
ratings_group_more_than_100.rename(columns={'rating': 'rating_count'}, inplace=True)
ratings_group_more_than_100.head()
```

```python
# Получаем максимальные значения timestamp по каждому userid
rating_max=ratings.groupby('userId').max().reset_index()
rating_max_timestamp=rating_max[['userId','timestamp']]
rating_max_timestamp.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
rating_max_timestamp.head()
```

```python
#Форматируем дату
rating_max_timestamp['date_max'] = pd.to_datetime(rating_max_timestamp['timestamp_max'], unit='s')
rating_max_timestamp.head()
```

```python
# Получаем минимальные значения timestamp по каждому userid
rating_min=ratings.groupby('userId').min().reset_index()
rating_min_timestamp=rating_min[['userId','timestamp']]
rating_min_timestamp.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
rating_min_timestamp.head()
```

```python
#Форматируем дату

rating_min_timestamp['date_min'] = pd.to_datetime(rating_min_timestamp['timestamp_min'], unit='s')
rating_min_timestamp.head()
```

```python
# Объединяем таблицы
avg_date=ratings_group_more_than_100.merge(rating_max_timestamp, how='inner').merge(rating_min_timestamp, how='inner')
```

```python
# Функция для нахождения времени жизни пользователя
def average(row):
    mean_=(row['date_max']-row['date_min'])
    return mean_
```

```python
# Применяем функцию к фрейму
avg_date['lifetime']=avg_date.apply(average, axis=1)
```

```python
avg_date.head(15)
```

```python
# Находим среднее значение  времени жизни пользователей, которые выставили более 100 оценок
lifetime_average_level = avg_date['lifetime'].mean()
lifetime_average_level
```

Задание 2

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:

таблицу с тремя типами выручки для каждого client_id без указания адреса клиента

аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table_rzd_auto = rzd.merge(auto, how='outer', on='client_id')
table_rzd_auto
```

```python
# таблицa с тремя типами выручки для каждого client_id без указания адреса клиента
table_rzd_auto_air=table_rzd_auto.merge(air, how='outer', on='client_id')
table_rzd_auto_air
```

```python
# аналогичная таблица по типам выручки с указанием адреса клиента
full_table = table_rzd_auto_air.merge(client_base, how='outer', on='client_id')
full_table
```
