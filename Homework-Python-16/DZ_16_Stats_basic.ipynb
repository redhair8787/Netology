```python
# Чтобы судить о состоянии здоровья лошадей, испытывающих кишечные колики, понадобятся столбцы hospital_number,  rectal temperature
# pulse, mucous membranes, abdominal distension, rectal examination, packed cell volume, abdomcentesis total protein ,
# outcome

#Отфильтруем датафрейм
horse_data_analysis = horse.filter(items = ['530101', '38.50', '66',  '4.1', '3.2', '45.00','8.40', '2.2'])


horse_data_analysis.columns = ['Номер больной лошади', 'Температура', 'Пульс',  'Вздутие живота', 
                               'Ректальное исследование', 'Количество эритроцитов в крови', 
                               'Общий белок', 'Результат']
horse_data_analysis
```

```python
# Расчет базовых метрик
stats_for_basics_metrics = ['Температура', 'Пульс', 'Количество эритроцитов в крови', 
                               'Общий белок']
for stats in stats_for_basics_metrics:
    print(f"Базовые метрики для {stats}")
    print('Размах', horse_data_analysis[stats].max() - horse_data_analysis[stats].min())
    print('Мода', horse_data_analysis[stats].mode()[0])
    print('Медиана', horse_data_analysis[stats].median())
    print('Дисперсия', horse_data_analysis[stats].var())
    print('Метод describe\n', horse_data_analysis[stats].describe())
    print('---------------------')
```

```python
#Размах для показателя "Пульс" на два порядка больше медианы, что указывает на сильный разброс данных. 
# Такая же тенденция наблюдается по показателю "Общий белок"
```

Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. 

Принять и обосновать решение о дальнейшей работе с ними.

```python
for stats in stats_for_basics_metrics:
    q1 = horse_data_analysis[stats].quantile(0.25)
    q3 = horse_data_analysis[stats].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    outliers = horse_data_analysis.loc[(horse_data_analysis[stats] < lower_bound)|
                                       (horse_data_analysis[stats] > upper_bound), stats]
    print(stats)
    print(outliers)
    print(f'Количество выбросов {outliers.count()}')
    print('--------------------')
```

Анализ выбросов:

    Выбросы по показателям "Температура", "Количество эритроцитов в крови" не нужно удалять или игнорировать, так как величины соответствуют температуре и показателю крови при болезни животного
    Выбросы по показателю "Пульс" нужно уточнять - возможен ли такой высокий пульс при шоковом состоянии и слишком низкий при начале болезни.
  
    

Задание 3. Работа с пропусками
Рассчитать количество выбросов для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for stats in horse_data_analysis.columns:
    pct_missing = horse_data_analysis[stats].isnull().mean()
    print(f'{stats} - {pct_missing :.1%}')
```

```python
#Температура - довольно много пропусков. Можно заменить медианным значением, т.к. размах по этой величине небольшой
horse_data_analysis_new = horse_data_analysis

horse_data_analysis_new['Температура'].fillna(horse_data_analysis['Температура'].median(), inplace = True)

#Пульс - можно заполнить пропуски средними значениями, тк оно ближе к норме
horse_data_analysis_new['Пульс'].fillna(horse_data_analysis['Пульс'].mode()[0], inplace = True)

#Вздутие живота - будем заполнять медианами для соответствующих значений в показателе "Результат" так как здесь
#наиболее полные данные и чем больше вздутие тем хуже состояние лошади
horse_data_analysis_new['Вздутие живота'].fillna(horse_data_analysis.groupby(['Результат'])['Вздутие живота'].transform('median'), inplace=True)

#Ректальные исследования - будем заполнять медианами для соответствующих значений в показателе "Результат" 
#и "Вздутие живота"
horse_data_analysis_new['Ректальное исследование'].fillna(horse_data_analysis.groupby(['Результат', 'Вздутие живота'])['Ректальное исследование'].transform('median'), inplace=True)

#Количество эритроцитов в крови и Общий белок будем заполнять медианами для соответствующих значений в "Результат". 
# Чем выше белок и больше эритроцитов, тем хуже состояние лошади.
horse_data_analysis_new['Количество эритроцитов в крови'].fillna(horse_data_analysis.groupby(['Результат'])['Количество эритроцитов в крови'].transform('median'), inplace=True)
horse_data_analysis_new['Общий белок'].fillna(horse_data_analysis.groupby(['Результат'])['Общий белок'].transform('median'), inplace=True)

#Результат - всего один пропуск - берем моду
horse_data_analysis_new['Результат'].fillna(horse_data_analysis['Результат'].mode()[0], inplace = True)

horse_data_analysis_new.head(40)
```
