```python
# Загрузка нужных библиотек
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
# Возьмем датасет с цветками iris’а
iris_flowers = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
iris_flowers.head()
```

```python

print(load_iris().DESCR)
```

```python
# Оставляем два признака - sepal_length и sepal_width и целевую переменную - variety

iris_flowers.drop(['petal length (cm)', 'petal width (cm)'], axis=1, inplace=True)
iris_flowers['variety'] = load_iris().target
iris_flowers
```

```python
# Разделяем данные на выборку для обучения и тестирования
from sklearn.model_selection import train_test_split
iris_flowers_train, iris_flowers_test = train_test_split(iris_flowers, test_size=0.2)
```

```python
# Получаем данные для обучения
iris_flowers_train
```

```python
# Получаем данные для тестирования
iris_flowers_test
```

```python
# Cтроим модель LDA
y_train, y_test = iris_flowers_train['variety'], iris_flowers_test['variety']
del iris_flowers_train['variety']
del iris_flowers_test['variety']
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(iris_flowers_train, y_train)
```

```python
result = pd.DataFrame([y_test.values, lda.predict(iris_flowers_test)]).T
result
```

```python
# Визуализируем предсказания для тестовой выборки и центры классов
plt.scatter(iris_flowers_train['sepal length (cm)'], iris_flowers_train['sepal width (cm)'], c=y_train)
plt.scatter(lda.means_[:, 0], lda.means_[:, 1], c='r', s=150, marker='*')
plt.show()
#визуализация для тринировочной выборки
```

```python
#для тестовой выюорки
plt.scatter(iris_flowers_test['sepal length (cm)'], iris_flowers_test['sepal width (cm)'], c=y_test)
plt.scatter(lda.means_[:, 0], lda.means_[:, 1], c='r', s=150, marker='*')
plt.show()
```

```python
# Отбросываем целевую переменную и оставьте только два признака - sepal_length и sepal_width
iris_flowers_2 = iris_flowers[['sepal length (cm)','sepal width (cm)']]
iris_flowers_2.columns = ['sepal_length','sepal_width']
iris_flowers_2.head()
```

```python
# Подбираем оптимальное число кластеров для алгоритма kmeans и визуализируем полученную кластеризацию
scaler = StandardScaler()
iris_flowers_2_scaled = scaler.fit_transform(iris_flowers_2)
plt.scatter(iris_flowers_2_scaled[:, 0], iris_flowers_2_scaled[:, 1])
```

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(iris_flowers_2_scaled)
    k_inertia.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_inertia)
plt.plot(ks, k_inertia ,'ro')

```

```python
#На мой взгляд, лучше разделить на 4 кластера
aglo = AgglomerativeClustering(n_clusters=4)
clusters = aglo.fit_predict(iris_flowers_2_scaled)
plt.scatter(iris_flowers_2_scaled[:, 0], iris_flowers_2_scaled[:, 1],cmap='spring', c=clusters, s=60)
```
