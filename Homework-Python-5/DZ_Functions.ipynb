```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# функция, которая выводит текущий список полок
def list_shelves(shelves):
    dir_keys = (",".join(shelves.keys()))
    return(dir_keys)



# Задаем функцию, которая запрашивает номер документа и по нему выводит имя (команда 'p')
     
def name_for_num_doc(list_of_docs):
    user_num_doc = input('Введите номер документа: ')
    for i in list_of_docs:
        
        #Если такой ключ в словаре существует и значение ключа равно номеру вводимого документа
        if 'number' in i.keys() and i['number'] == user_num_doc:
            name = i['name']
            return(f'Владелец документа:  {name}')
            
    return 'Документ не найден в базе'

# print(name_for_num_doc(documents))
 
# Задаем функцию, которая по номеру документа выводит номер полки (команда 's')    
def shelf_for_num_doc(shelves, s_user_num_doc = None):
    var_exist = False
    if s_user_num_doc is None:
        user_num_doc = input('Введите номер документа: ')
    else:
        user_num_doc = s_user_num_doc
        var_exist = True
        
    for i in shelves:
        if user_num_doc in shelves[i]:
            if var_exist == False:
                return(f'Документ хранится на полке: {i}')
            else:
                return(i)
    return 'Документ не найден в базе'
# print(shelf_for_num_doc(directories))

# Задаем функцию, которая выводит полную информацию по всем документам (команда 'l')
def information_docs(list_of_docs):
    for i in list_of_docs:
        l_number = i['number']
        l_type = i['type']
        l_name = i['name']
        l_shelf = shelf_for_num_doc(directories, i['number'])
        print(f'№: {l_number}, тип: {l_type}, владелец: {l_name}, полка хранения: {l_shelf}')
        
    return 0
# print(information_docs(documents))

# Задаем функцию, которая добавляет новую полку (команда 'as')
def new_shelf(shelves):
    user_num_shelf = input('Введите номер полки')
    if user_num_shelf in shelves:
        dir_keys = (",".join(shelves.keys()))
        return(f'Такая полка уже существует. Текущий перечень полок: {dir_keys}') 
    shelves[user_num_shelf] = []
    
    dir_keys = (",".join(shelves.keys()))
    return(f'Полка добавлена. Текущий перечень полок: {dir_keys}')
# print(new_shelf(directories)) 


# Задаем функцию, которая удаляет существующую полку из данных (только если она пустая) - команда 'ds'

def delete_shelf(shelves):
    
    user_num_shelf = input('Введите номер полки')
    if shelves.get(user_num_shelf):
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list_shelves(shelves)}')
    # если полка существует и она пустая
    elif user_num_shelf in shelves:
        del shelves[user_num_shelf]
        return(f'Полка удалена. Текущий перечень полок: {list_shelves(shelves)}')

    else:
        
        return(f'Такой полки не существует. Текущий перечень полок: {list_shelves(shelves)}')
# print(delete_shelf(directories)) 


# Функция, которая по введенным командам вызывает соответствующие функции 
def main(list_of_docs, shelves):
    while True:
        user_command = input('Введите команду ')
        if user_command == 'p':
            print(name_for_num_doc(documents))
        elif user_command == 's':
            print(shelf_for_num_doc(directories))
        elif user_command == 'l':
            information_docs(documents)
        elif user_command == 'as':
            print(new_shelf(directories))
        elif user_command == 'ds':
            print(delete_shelf(directories))
        elif user_command == 'q':
            print('Завершение программы!')
            break

main(documents, directories)

  
```
